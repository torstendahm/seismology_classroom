# Fourier Transformation

Whats done here

```python
import numpy as num
import matplotlib.pyplot as plt
```

```python
t = num.arange(1000)
fac = 10
x = num.sin(t/fac)
#x = num.sin(t/fac) + num.sin(2*t/fac)
#x = num.sin(t/fac) + num.sin(2*t/fac) + num.sin(10*t/fac)



plt.figure()
plt.plot(t, x)
plt.show()
```

```python


sp = num.fft.rfft(x)
freq = num.fft.rfftfreq(t.shape[-1])

plt.figure()
plt.semilogx(freq, abs(sp))
plt.show()
```
