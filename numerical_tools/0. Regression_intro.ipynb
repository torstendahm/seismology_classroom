# 1D Regression Analysis

Whats is done
- Least square fit

```python
import numpy as num
import matplotlib.pyplot as plt
```

```python
# Create data

xdata = num.linspace(-100, 100, 1000)

num.random.seed(0)
noise = num.random.normal(0, 100, len(xdata))
#print(noise)
ydata = 5 * xdata + noise

plt.figure()
plt.scatter(xdata, ydata)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
xdata = xdata.reshape((-1, 1))

model.fit(xdata, ydata)

r_sq = model.score(xdata, ydata)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
yregr = model.coef_ * xdata + model.intercept_
# yregr = model.predict(xdata)

plt.figure()
plt.scatter(xdata, ydata, label='Data')
plt.plot(xdata, yregr, color='red', label='Regression')
plt.legend()
plt.show()
```

```python
# Prediction of new samples
## PREDICTION OUTSIDE OF REGRESSION LEARNING DATA should be avoided

new_xdata = num.linspace(-100, 0, 25).reshape((-1, 1))

ypred = model.predict(new_xdata)


plt.figure()
plt.scatter(new_xdata, ypred)
plt.show()
```

# With numpy polynomial

```python
# Create data

xdata = num.linspace(-100, 100, 1000)

num.random.seed(0)
noise = num.random.normal(0, 100, len(xdata))
#print(noise)
ydata = xdata * 1 + noise
#ydata = num.sin(xdata * 2 * num.pi * 0.01) + xdata * 0.01


order = 5

coeff = num.polyfit(xdata, ydata, order)
print('Coeffs:', coeff)

mymodel = num.poly1d(coeff)


plt.figure()
plt.scatter(xdata, ydata, label='Data')
plt.plot(xdata, mymodel(xdata), color='red', label='Regression')
plt.legend()
plt.show()
```
