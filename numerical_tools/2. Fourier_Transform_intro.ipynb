# Fourier Transformation

Whats done here

Amplitude and phase spectrum 

```python
import numpy as num
import scipy as sp
import matplotlib.pyplot as plt
```

```python
df = 100  # Abstastfrequenz [Hz]
timelen = 10  # LÃ¤nger der Zeitspur [s]
dt = 1 / df  # Abtastintervall [s]

numsamples = timelen * df
x = num.arange(numsamples) / df
```

```python
sin1 = 1.
ysignal = num.sin(2 * num.pi * sin1 * x)

amps = abs(num.fft.rfft(ysignal)) * dt
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(2, 1, 2)
ax.plot(freqs, amps)
# ax.loglog(freqs, amps)

plt.show()
```

```python
sin1 = 10.
ysignal = num.sin(2 * num.pi * sin1 * x)

amps = abs(num.fft.rfft(ysignal)) * dt
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(2, 1, 2)
ax.plot(freqs, amps)
# ax.loglog(freqs, amps)

plt.show()
```

```python
sin1 = 1.
sin2 = 10.
ysignal = num.sin(2 * num.pi * sin1 * x) + num.sin(2 * num.pi * sin2 * x)

amps = abs(num.fft.rfft(ysignal)) * dt
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(2, 1, 2)
# ax.plot(freqs, amps)
ax.loglog(freqs, amps)

plt.show()
```

```python
sin1 = 1.
cos1 = 10.
ysignal = num.sin(2 * num.pi * sin1 * x) + num.cos(2 * num.pi * cos1 * x)

amps = abs(num.fft.rfft(ysignal)) * dt
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(2, 1, 2)
# ax.plot(freqs, amps)
ax.loglog(freqs, amps)

plt.show()
```

```python
frequencies = num.linspace(1, 100, 100)
# print(frequencies)

ysignal = num.zeros(numsamples)
for ff in frequencies:
    ysignal += (1/ff) * num.sin(2 * num.pi * ff * x)
    #ysignal += ff * num.sin(2 * num.pi * ff * x)

amps = abs(num.fft.rfft(ysignal)) * dt
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(2, 1, 2)
ax.plot(freqs, amps)
# ax.loglog(freqs, amps)

plt.show()
```
