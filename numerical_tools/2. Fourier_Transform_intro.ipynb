```python
import numpy as num
import scipy as sp
import matplotlib.pyplot as plt
```

```python
df = 100  # Abstastfrequenz [Hz]
timelen = 10  # LÃ¤nger der Zeitspur [s]
dt = 1 / df  # Abtastintervall [s]

numsamples = timelen * df
x = num.arange(numsamples) / df
```

```python
sin1 = 1 # frequency of signal, inverse of periode
ysignal = num.sin(2 * num.pi * sin1 * x) + num.random.normal(0, 0.1, len(x))

plt.figure()
plt.plot(x, ysignal)
plt.grid()
plt.show()
```

```python
ft = num.fft.rfft(ysignal)

amps = abs(ft) * dt
phase =  num.angle(ft)
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure()
#plt.plot(freqs, amps)
plt.semilogx(freqs, amps)
#plt.loglog(freqs, amps)
plt.show()

#plt.figure()
#plt.plot(freqs, phase / num.pi)
#plt.show()

#plt.figure()
#plt.phase_spectrum(ysignal, Fs=1/dt)
#plt.show()
```

```python
sin1 = 10.
ysignal = num.sin(2 * num.pi * sin1 * x)

amps = abs(num.fft.rfft(ysignal)) * dt
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(2, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(2, 1, 2)
ax.plot(freqs, amps)
#ax.semilogx(freqs, amps)
#ax.loglog(freqs, amps)

plt.show()

## play with different frequencies 0.1 till nyquist
```

# Multiple Frequencies
## plus Phase

```python
sin1 = 1.
sin2 = 10.
ysignal = num.sin(2 * num.pi * sin1 * x) + num.sin(2 * num.pi * sin2 * x)

ft = num.fft.rfft(ysignal)

amps = abs(ft) * dt
phase =  num.angle(ft)
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(3, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(3, 1, 2)
ax.plot(freqs, amps)
#ax.semilogx(freqs, amps)
#ax.loglog(freqs, amps)
ax = plt.subplot(3, 1, 3)
ax.plot(freqs, phase)

plt.show()
```

```python
sin1 = 1.
cos1 = 10.
ysignal = num.sin(2 * num.pi * sin1 * x) + num.cos(2 * num.pi * cos1 * x)

ft = num.fft.rfft(ysignal)

amps = abs(ft) * dt
phase =  num.angle(ft)
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(3, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(3, 1, 2)
ax.plot(freqs, amps)
#ax.semilogx(freqs, amps)
#ax.loglog(freqs, amps)
ax = plt.subplot(3, 1, 3)
ax.plot(freqs, phase)

plt.show()
# slightly different waveform, same amp spectrum
```

```python
frequencies = num.linspace(1, 100, 100)
# print(frequencies)

ysignal = num.zeros(numsamples)
for ff in frequencies:
    ysignal += (1/ff) * num.sin(2 * num.pi * ff * x)
    #ysignal += ff * num.sin(2 * num.pi * ff * x)

ft = num.fft.rfft(ysignal)

amps = abs(ft) * dt
phase =  num.angle(ft)
freqs = num.fft.rfftfreq(numsamples, d=dt)

plt.figure(figsize=(16, 9))
ax = plt.subplot(3, 1, 1)
ax.plot(x, ysignal)
ax = plt.subplot(3, 1, 2)
ax.plot(freqs, amps)
#ax.semilogx(freqs, amps)
#ax.loglog(freqs, amps)
ax = plt.subplot(3, 1, 3)
ax.plot(freqs, phase)

plt.show()
```
