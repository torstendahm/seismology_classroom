```python
import numpy as num
import matplotlib.pyplot as plt
x = num.linspace(0, 10, 100)
y = 10**x

plt.figure()
plt.plot(x, y)
#plt.yscale('log')
plt.show()

plt.figure()
plt.plot(x, num.log(y))
plt.show()
```
