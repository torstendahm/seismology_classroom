# Coordinate systems of seismograms

## Coordinate systems used for seismogram recording

Seismic waves are recorded using seismometers. Besides special sensors (e.g. rotational seismometers) seismometers record the unidirectional ground motion in the 3 dimensions. Each indivdual record in one dimension is called a **channel** or **component** of the seismometer. Each individual channel is clearly defined by its name. Below an example for a 3D record is shown from the 2020 M<sub>w</sub> 7.0 Samos earthquake:

Different channel setups are commonly used within seismology for recording sensors:

| Coordinate system | Positive direction of components | Sketch (map view) &copy; IRIS|
|:---:|:---|:---:|
|__ZNE__| <ul><li>Z - upwards</li><li>N - north</li><li>E - east</li></ul> | <img width=300 src="https://ds.iris.edu/media/webservicedoc/irisws/rotation/1/zne.png"> |
|__Z12__| <ul><li>Z - upwards</li><li>1 - randomly oriented horizontal</li><li>2 - horizontal and rotated 90 degrees clockwise from 1 </li></ul> |<img width=300 src="https://ds.iris.edu/media/webservicedoc/irisws/rotation/1/z12.png"> |
 
<div class="alert alert-warning">
    <p style="font-weight: bold; font-size: 150%">Take care:</p>
    <ol>
        <li>Orientations of the different components are (often) accurate for permanently installed seismometers.</li>
        <li>Seismic stations of temporal networks might have strong orientation errors especially on horizontal components (even when called N or E).</li>
    </ol>
</div>


## Coordinate systems used for seismogram analysis

Besides the mentioned **ZNE** coordinate system other component orientations are used in the analysis of seismograms. They allow for a better assessment of different phases of the seismic wave, as they seperate different phases onto the different components (in theory).

Why? Because the phases (as shown in [1.0 - Introduction to Picking](1.0_introduction.ipynb)) are characterized by rather different waveform excitations. P-waves, for example, show compressional and dilatational motion in direction of waveform propagation, while S-waves are characterized by shearing perpendicular to the travel path of the waves.

The used coordinate systems are:

| Coordinate system | Positive direction of components | Sketch &copy; IRIS|
|:---:|:---|:---:|
|__ZRT__| <ul><li>Z - upwards</li><li>R - radial (parallel to wave propagation direction)</li><li>T - transversal (perpendicular to wave propagation direction)</li></ul> | <img width=300 src="https://ds.iris.edu/media/webservicedoc/irisws/rotation/1/zrt.png"> |
|__LQT__| <ul><li>L - Latitudinal</li><li>Q - SV-wave shearing direction</li><li>T - SH-wave shearing direction </li></ul> |<img width=300 src="https://ds.iris.edu/media/webservicedoc/irisws/rotation/1/lqt.png"> |

Waveforms recorded in **ZNE** are rotated into either **ZRT** or **LQT** and then further analysed.
